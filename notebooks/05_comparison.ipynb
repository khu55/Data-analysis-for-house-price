```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
results = pd.DataFrame({
    "Model": [
        "Linear Regression",
        "Random Forest",
        "PyTorch MLP"
    ],
    "RMSE": [
        37935.34,
        24530.32,
        31812.84
    ],
    "MAE": [
        21688.04,
        14842.23,
        17616.91
    ],
    "R2": [
        0.8205,
        0.92495,
        0.87377
    ]
})

results
```

```python
results.set_index("Model")["RMSE"].plot(
    kind="bar",
    figsize=(8,5),
    title="Model Comparison by RMSE"
)

plt.ylabel("RMSE")
plt.show()
```
